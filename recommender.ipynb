```python
content["team_code"] = content["Team"].astype("category").cat.codes
```

```python
content["player_code"] = content["Player"].astype("category").cat.codes
```

```python
unique_seasons = sorted(content["Season"].unique(), key= lambda x: int(x.split('-')[0]))
content["Season"] = pd.Categorical(content["Season"], categories=unique_seasons, ordered=True)
content["Season_code"] = content["Season"].astype("category").cat.codes.astype("int")
```

```python
from sklearn.metrics import f1_score, precision_score
```

```python
def validate_classifier(model, content, predictors, targets):
    seasons = content["Season_code"].unique()[::-1]
    f1_scores = {target: [] for target in targets}
    precision_scores = {target: [] for target in targets}
    for season in seasons[1:]:
        test = content[content["Season_code"] == season]
        train = content[content["Season_code"] < season]
        model.fit(train[predictors], train[targets])
        preds = model.predict(test[predictors])
        actual = test[targets]

        for i, target in enumerate(targets):
            f1 = f1_score(actual[target], preds[:,i], zero_division=0)
            precision = precision_score(actual[target],preds[:,i], zero_division=0) 
            f1_scores[target].append(f1)
            precision_scores[target].append(precision)
        # print(f"Season {season} F1 Score: {f1}")
    return {
        f"{target}_F1": sum(f1_scores[target]) / len(f1_scores[target])
        for target in targets
    } | {
        f"{target}_Precision": sum(precision_scores[target]) / len(precision_scores[target])
        for target in targets
    }


```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python

dt = DecisionTreeClassifier(class_weight="balanced")
```

```python
predictors = [ "Age", "GP", "W", "L", "Min", "PTS", "FGM", "FGA", "FG%",
    "3PM", "3PA", "3P%", "FTM", "FTA", "FT%", "OREB", "DREB", "REB",
    "AST", "TOV", "STL", "BLK", "PF", "DD2", "TD3", "+/-", "team_code",
]
targets = ["MVP", "DPOY", "ROTY", "6MOTY"]
```

```python
metrics = validate_classifier(dt, content, predictors, targets)
print("Decision Tree Metrics")
for name, value in metrics.items():
    print(f"{name}: {value}")
```

Using Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.multioutput import MultiOutputClassifier
```

```python
rf = MultiOutputClassifier(RandomForestClassifier(class_weight="balanced", random_state=42))
rf_metrics = validate_classifier(rf, content, predictors, targets)
print("Random Forest Metrics")
for name, value in rf_metrics.items():
    print(f"{name}: {value}")
```

```python
from sklearn.dummy import DummyClassifier
dummy = DummyClassifier(strategy='most_frequent')
dummy_metrics = validate_classifier(dummy, content, predictors, targets)
print("Dummy Classifier Metrics")
for name, value in dummy_metrics.items():
    print(f"{name}: {value}")
```
